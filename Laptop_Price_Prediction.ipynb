```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/asrulabdullah99/ml_dataset/main/laptop_price.csv"
df = pd.read_csv(url, encoding='latin1')
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
df_sorted = df.sort_values(by="Price_euros")

plt.figure(figsize=(10,5))
plt.bar(range(len(df_sorted)), df_sorted["Price_euros"])
plt.title("Harga Laptop (Ascending)")
plt.show()
```
